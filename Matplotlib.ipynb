# line graph

```python

a = [1,2,3,4,5] #x axis
b = [1,4,9,16,25] # y axis

c = [1,8,27,64,125] 

plt.plot(a,b, color='red', label='Squares', linestyle='dashed', marker='o', markersize=12, markerfacecolor='blue')
plt.plot(a,c, color='green', label='Cubes', linestyle='-.')
plt.xlabel('Numbers')
plt.ylabel('Squares and Cubes')

plt.title('Squares and Cubes of numbers')
plt.legend()



```

```python
a = [1,2,3,4]
b = [1,4,9,16]
c = [1,8,27,64]

# first line color, markershape, line style, marker size , marker color , loc to were locate the legend, mec to colour the outline 
# of the marker (marker edge color)
plt.plot(a,b,'ys-',label='squares',markersize = 10, markerfacecolor = 'purple') # a on x-axis and b on y-axis 
plt.plot(a,c,'gD--',label='cubes',markersize = 20, markerfacecolor = 'green', mec ='red') # a on x-axis and c on y-axis 
plt.title('squares and cube roots')
plt.xlabel('Numbers')
plt.ylabel('squares and cubes')
plt.legend(loc = 0)

plt.grid()
plt.savefig('save of number practice .png')
```

```python
a = [1,2,3,4]
b = [1,4,9,16]
c = [1,8,27,64]

plt.figure(figsize=(10,10))
plt.plot(a,b, 'cD--', markersize=20, markerfacecolor='red',mec='blue', label='Squares')

plt.plot(a,c, 'mH-.', markersize=20, markerfacecolor='blue',mec='red',label = 'Cubes',linewidth = '8')
plt.xlabel('Numbers',color='blue',fontsize=20)
plt.ylabel('Squares and cubes',color='blue',fontsize=20)
plt.title('Squares root and cube roots',color='red',fontsize=20)
plt.legend(fontsize = 20, loc=0)

plt.grid()
# other color is hotpink 
```

```python
fruits = ['Apple', 'Banana', 'Chiku', 'Mango']
price1day  = [80,100,20,70]
price2day= [34,90,18,60]


plt.figure(figsize=(18,8))


index = np.arange(len(fruits))
width = 0.5

plt.title('colBar Graph',fontsize=30, color='red')

plt.xlabel('fruits', fontsize=30, color='red')
plt.ylabel('Fruits kg and price',fontsize=30, color='red')

plt.bar(index, price1day, width, color='blue', label='price1day')
plt.bar(index + width, price2day, width, color='green', label='price2day')

plt.legend(fontsize=20)
```

```python
# total fruits sold 
fruits = ['Apple', 'Banana', 'Chiku', 'Mango']
price1day  = [80,100,20,70]
price2day= [34,90,18,60]

plt.figure(figsize=(18,8))

index = np.arange(len(fruits))
width = 0.5

plt.title('Graph for total numbers of fruits sold',fontsize=30, color='red')

plt.xlabel('fruits', fontsize=30, color='red')
plt.ylabel('Fruits kg and price',fontsize=30, color='red')

plt.bar(index, price1day, width, color='blue', label='price1day')
plt.bar(index, price2day, width, color='green', label='price2day', bottom = price1day)
plt.xticks(index,fruits, fontsize= 20)
plt.legend(fontsize=20)
```
