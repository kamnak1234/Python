# Combine a dataset

```python
a = [['A',10],['B',20],['C',30]]
b1 = pd.DataFrame(a, columns=['x1','x2']) 
print(b1)
```

```python
a = [['A','T'],['B','F'],['D','T']]
b2 = pd.DataFrame(a, columns=['x1','x3']) 
print(b2)
```

```python
pd.merge(b1,b2 ,how='left', on='x1') # Join matching rows from b2 to b1
```

```python
pd.merge(b1,b2 ,how='right', on='x1') #Join matching rows from b1 b2.
```

```python
pd.merge(b1,b2 ,how='inner', on='x1')#Join data. Retain only rows in both sets.
```

```python
pd.merge(b1,b2 ,how='outer', on='x1')#Join data. Retain all values, all rows.
```

```python
b1[b1.x1.isin(b2.x1)] #All rows in b1 that have a match in b2.
```

```python
b1[~b1.x1.isin(b2.x1)]#All rows in b1 that do not have a match in b2.
```

```python
a = [['A',1],['B',2],['C',3]]
y = pd.DataFrame(a, columns=['x1','x2']) 
print(y)
```

```python
a = [['B',2],['C',3],['D',4]]
z = pd.DataFrame(a, columns=['x1','x2']) 
print(z)
```

```python
pd.merge(y, z) #Rows that appear in both ydfand zdf (Intersection)
```

```python
pd.merge(y, z, how='outer') #Rows that appear in either or both ydfand zdf (Union).
```

```python
pd.merge(y,z,how='outer',indicator=True).query('_merge == "left_only"').drop(columns=['_merge']) 
#Rows that appear in ydfbut not zdf(Setdiff)
```
