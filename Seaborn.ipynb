# C.Distribution of data (sns.displot.sns.jointplot,sns.pairplot)

```python
sns.displot(a['age'])
```

```python
#increases band width
sns.displot(a['age'],binwidth=10)
```

```python
#increases the bins
sns.displot(a['age'],bins=20)
```

```python
#curve
sns.displot(a['age'],kind='kde')
```

```python
#curve and bar both
sns.displot(a['age'], kde = True, rug=True)
```

```python
rug and bins 
sns.displot(a['age'], kde = True, rug=True,bins=90)
```

```python
#discrete avoid the gaps 
sns.displot(a, x="age", discrete=True)
```

```python
#shrink to add thye gap as per passing value 
sns.displot(a, x="age",shrink=.8)
  
```

```python
#hue will show the overlap bar 
sns.displot(a, x="age",shrink=.8,hue='sex')
```

```python
# element to avoid the overlap
sns.displot(a, x="age",shrink=.8,hue='sex',element='step')
```

```python
#to overlapp the multiple data 
sns.displot(a, x="age",shrink=.8,hue='sex',multiple="stack")
```

```python
#multiple data one after another 
sns.displot(a, x="age",shrink=.8,hue='sex',multiple="dodge")
```

```python
sns.displot(a, x="age", hue="sex", kind="kde", multiple="stack")
```
